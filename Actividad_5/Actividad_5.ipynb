```python
#Vemos valores nulos de ginebra

valores_nulos_ginebra = cs.identificacion_nulos(df_ginebra)
```

```python
#Generamos una copia de ginebra y mexico

df1_ginebra = df_ginebra.copy()
df1_mexico = df_mexico.copy()
```

```python
#Reemplazamos valores nulos de ginebra

df2_ginebra = cs.sustitucion_ffill(df1_ginebra)
df3_ginebra = cs.sustitucion_mean(df2_ginebra)
nulos_ginebra = cs.identificacion_nulos(df3_ginebra)
```

```python
#Reemplazamos valores nulos de mexico

df2_mexico = cs.sustitucion_ffill(df1_mexico)
df3_mexico = cs.sustitucion_mean(df2_mexico)
nulos_mexico = cs.identificacion_nulos(df3_mexico)
```

```python
#Creamos un dataframe de ginebra para las cualitativas y otro para las cuantitativas (agregamnos datetime para las numericas)

cuantitativas_ginebra = df3_ginebra.select_dtypes(include=['int64', 'float64','datetime64[ns]'])
cualitativas_ginebra = df3_ginebra.select_dtypes(include=['object'])
```

```python
#Creamos un dataframe de mexico para las cualitativas y otro para las cuantitativas

cuantitativas_mexico = df3_mexico.select_dtypes(include=['int64', 'float64'])
cualitativas_mexico = df3_mexico.select_dtypes(include=['object'])
```

```python
#Realizamos diagrama de caja o bigote de cada columa del datarame de ginebra

fig_ginebra = plt.figure(figsize=(15, 8))
cuantitativas_ginebra.plot(kind='box', vert=False)
plt.title('Valores Atípicos de Ginebra')
plt.show() #dibujamos el diagrama
```

```python
#Realizamos diagrama de caja o bigote de cada columa del datarame de mexico

fig_mexico = plt.figure(figsize=(15, 8))
cuantitativas_mexico.plot(kind='box', vert=False)
plt.title('Valores Atípicos de Ginebra')
plt.show() #dibujamos el diagrama
```

PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLINERS EN DATAFRAMES

```python
#Método aplicando Cuantiles. Encuentro cuartiles 0.25 y 0.75 para ginebra
y_ginebra = cuantitativas_ginebra

percenttile25_ginebra = y_ginebra.quantile(0.25) #Q1
percenttile75_ginebra = y_ginebra.quantile(0.75) #Q3
iqr_ginebra = percenttile75_ginebra - percenttile25_ginebra #Rango intercuartil #cajita

Limite_Superior_iqr_ginebra = percenttile75_ginebra + 1.5 * iqr_ginebra
Limite_Inferior_iqr_ginebra = percenttile25_ginebra - 1.5 * iqr_ginebra
print("Limite superior permitido", Limite_Superior_iqr_ginebra)
print("Limite inferior permitido", Limite_Inferior_iqr_ginebra)
```

```python
#Método aplicando Cuantiles. Encuentro cuartiles 0.25 y 0.75 para mexico
y_mexico = cuantitativas_mexico

percenttile25_mexico = y_mexico.quantile(0.25) #Q1
percenttile75_mexico = y_mexico.quantile(0.75) #Q3
iqr_mexico = percenttile75_mexico - percenttile25_mexico #Rango intercuartil #cajita

Limite_Superior_iqr_mexico = percenttile75_mexico + 1.5 * iqr_mexico
Limite_Inferior_iqr_mexico = percenttile25_mexico - 1.5 * iqr_mexico
print("Limite superior permitido", Limite_Superior_iqr_mexico)
print("Limite inferior permitido", Limite_Inferior_iqr_mexico)
```

```python
#Obtenemos datos limpios del DataFrame ginebra
ginebra_iqr = cuantitativas_ginebra[(y_ginebra<Limite_Superior_iqr_ginebra) & (y_ginebra>Limite_Inferior_iqr_ginebra)]
ginebra_iqr
```

```python
#Obtenemos datos limpios del DataFrame mexico
mexico_iqr = cuantitativas_mexico[(y_mexico<Limite_Superior_iqr_mexico) & (y_mexico>Limite_Inferior_iqr_mexico)]
mexico_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO de ginebra
valores_nulos_ginebra_iqr = cs.identificacion_nulos(ginebra_iqr)
valores_nulos_ginebra_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO de mexico
valores_nulos_mexico_iqr = cs.identificacion_nulos(mexico_iqr)
valores_nulos_mexico_iqr
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe ginebra con "mean"
#Realizamos una copia del dataframe
ginebra_iqr2 = ginebra_iqr.copy()
ginebra_iqr2 = cs.sustitucion_mean(ginebra_iqr2)
nulos_ginebra_iqr2 = cs.identificacion_nulos(ginebra_iqr2)
```

```python
#Eliminamos columnas sin valores (ID SCRAPE, REVIEWS ULTIMOS 30D, ANUNCIOS HABITACION COMPARTIDA) y revisamos nulos de nuevo
ginebra_iqr2 = ginebra_iqr2.drop(['ID SCRAPE', 'REVIEWS ULTIMOS 30D', 'ANUNCIOS HABITACION COMPARTIDA'], axis=1)
nulos_ginebra_iqr2 = cs.identificacion_nulos(ginebra_iqr2)
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo para ginebra

Datos_limpios_ginebra = pd.concat([cualitativas_ginebra, ginebra_iqr2], axis=1)
Datos_limpios_ginebra
```

```python
#Corroboramos valores nulos del dataframe LIMPIO de ginebra
valores_nulos_ginebra_iqr2 = cs.identificacion_nulos(Datos_limpios_ginebra)
valores_nulos_ginebra_iqr2
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe mexico con "mean"
#Realizamos una copia del dataframe
mexico_iqr2 = mexico_iqr.copy()
mexico_iqr2 = cs.sustitucion_mean(mexico_iqr2)
nulos_mexico_iqr2 = cs.identificacion_nulos(mexico_iqr2)
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo para mexico

Datos_limpios_mexico = pd.concat([cualitativas_mexico, mexico_iqr2], axis=1)
Datos_limpios_mexico
```

```python
#Corroboramos valores nulos del dataframe LIMPIO de mexico
valores_nulos_mexico_iqr2 = cs.identificacion_nulos(Datos_limpios_mexico)
valores_nulos_mexico_iqr2
```
